# Missing data

In missing data you fill the data of remove the element , you want to remove the element , you want to create a algorithm you dont want to messy data in it

```python
import numpy as np
import pandas as pd
```

```python
# np.nan = null value

data = {
    'A':[1,2,3,4,5],
    'B':[np.nan,2,3,4,5],
    'C':[1,2,3,np.nan,np.nan],
    'D':[1,np.nan,np.nan,np.nan,5]
}

df = pd.DataFrame(data)
df
```

**Finding Missing Data**

```python
# .isna() check the null value in data if present output is True

df.isna() # show data which one is nan represents as True
```

```python
# for the count of nan value in Dataframe

df.isna().sum()
```

```python
# To check in column ( or Row ) is null value , use .any()

df.isna().any()
```

**Removing Missing Data**

```python
df.dropna() # use for removing the null value , works on row , but this removes whole row of null value if 1 nan is present
```

```python
# there comes a thresh you can set limit how much nan is allowed in df , You have Atleast NonValue of 3 (thresh=3)

df.dropna(thresh=3)
```

**Filling the missing value**

```python
# df.fillna() to fill the data is null value , it only apply on column

df.fillna(1) # it will be replace the Nan value to 1 , change will not happen in original dp ,
             #because we not using a ( inplace="True" ) only it can change a df Permentetly
```

```python
# you can do like this also if you want a specific column a specific values comes as you input , For a different values for each column
values = {'A':100,'B':200,'C':300,'D':400} # if there is no nan value in columnn , no values output will be there
```

```python
df.fillna(value=values)
```

```python
# you want to fill mean value on nan value

df.fillna(df.mean()) # it fill thw value of nan with mean value
```
