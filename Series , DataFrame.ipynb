# Series 
A Series is one-dimensional labeled array capable of holding any datatype this axis labels are collectively called the index

**Creating Series**

There are multiple ways to create a pandas series:
--> [1,2,3,4]    

series = Column

```python
import numpy as np
import pandas as pd
```

```python
labels = ['a','b','c']
my_list = [10,20,30]
arr = np.array(my_list) 
dictn = {
    "A":10,
    "B":20,
    "C":30
}
```

```python
# To create a series we use a .series()
pd.Series(my_list) # create a series , S should be capital letter
# in Series a output will be in vertically (labelsb)
```

```python
pd.Series(arr)
```

```python
pd.Series(dictn) # we use as series we automatically set index by use key : value pair
```

```python
pd.Series(my_list,index=labels) # custom series index , you can customize a index as your list as given here is labels
```

# Dataframe ( dictionaries , list )

when we combine a multiple series together it will become a Dataframe

```python
import numpy as np
import pandas as pd
```

**Creating a Dataframe**

```python
data = {
    'name':['john','anna','peter','linda'],
    'age':[28,34,29,42],
    'city':['new york','paris','berlin','london'],
    'salary':[65000,70000,62000,85000]    
}
```

```python
pd.DataFrame(data) # create a dataframe of data dict
```

```python
df = pd.DataFrame(data) # it will show df with title or column name , we store dataframe as df
df
```

```python
columns = ['name','age'] # we use it when we want some of columns in output
```

```python
df = pd.DataFrame(data,columns=columns)
df
```

**Selection and Indexing of columns**

```python
df['name'] # Selecting a columns from name 
```

```python
# for both or multiple column access use [[ex1,ex2]] list of list type

df[['name','age']]
```

**Creating a new column**

```python
# for creating a new column we create as basic like df["new_column_name"]=['val1','val2',...'valn']

df['designation']=['IT','HR','Doctor','CEO'] # added new column at df you can add with 4 value existing not less more
df
```

**Removing a column**

```python
# for dropping a column we use a .drop 
# df.drop('designation') it will show error like axis , cannot found is axis

df.drop('designation',axis=1) # because axis=0 works on rows you get error , remove only on copy or temporary but in df is present
```

```python
# that why here comes a inplace that drop a column a permantetly , inplace=True

df.drop('designation',axis=1,inplace=True) # Delete a permentetly Column
```

```python
# for drop a row also work

df.drop(0,axis=0) # it remove a row .drop(index_no,axis=1/0) 1 = column , 0 = row
```

**Selecting Rows**

```python
# for accessing or selecting the row .loc(label based)

df.loc[0] # it show a row = 0 value .loc(location)
```

```python
# for accessing a row and column 

df.loc[0:2,'name'] # as .loc[row,column]
```

```python
# for accessing a multiple rows [[val1,val2]] like list of list

df.loc[[1,3]] 
```

```python
# we can slice a value also we want particular value

df.loc[0:2] # Note: This returns rows 0, 1, and 2 (excludes index 3)
```

```python
# Boolean Indexing: df.loc[df['age'] > 30] — This is the most common way to filter data based on conditions.
# List of Labels: df.loc[['r1', 'r2'], ['c1', 'c2']]
```

```python
# i stand for index location .iloc(integer-location-based)
df.iloc[0] 

# Negative Indexing: df.iloc[-1] returns the very last row
```

**Selecting Subsets of rows and columns**

```python
# The general syntax is: df.loc[row_selection, column_selection]. 

df.loc[[0,1],['name','age']]
```

```python
# Pratice work :
df.loc[[2,3],['name','age']]
```

**Conditional Selection**

```python
# ( < , > , <= , >= , ! , != , == )

# Q1. I only want to see those people whose age is above 30 . syntax : df[df["Col"]>0]

df[df['age']>30]
```

```python
# I only want to see those people whose age is above 30 there city must be paris
df[(df['age']>30) & (df['city']=="paris")]
```
