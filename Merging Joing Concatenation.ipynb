```python
import numpy as np
import pandas as pd
```

```python
# DataFrame 1
employees = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'name':['John','Anna','Peter','Linda','Bob'],
    'department':['HR','IT','Finance','IT','HR']
})
```

```python
# DataFrame 2 
salaries = pd.DataFrame({
    'employee_id':[1,2,3,4,5],
    'salary':[60000,80000,65000,70000,90000],
    'bonus':[5000,10000,7000,8000,12000]
})
```

# Merging 2 Dataframe 

```python
# Using a .merge() // merges a DataFrame

pd.merge(employees,salaries) # it will merge the both dataframe
```

**on=**  If common columns is both dataframe you can use as a join

**how='inner'**  It give the inner value means the based on 1 employees

**how='outer'**  Gives all inner + outer value , mean both dataframe

**how='left'** We want merge basis on left(employee) i want all the id on that present on employees

**how='right'** if i do operation right basis it will perform on salaries

If there are empty values in dataframe it will showns as Nan

```python
pd.merge(employees,salaries,on='employee_id',how='inner')
```

```python
pd.merge(employees,salaries,on='employee_id',how='outer')
```

```python
pd.merge(employees,salaries,on='employee_id',how='left')
```

```python
pd.merge(employees,salaries,on='employee_id',how='right')
```

# Concatination of 2 dataframe

```python
df1 = pd.DataFrame({
    'A':['A0','A1','A2'],
    'B':['B0','B1','B2'],
    'C':['C0','C1','C2']
})
```

```python
df2 = pd.DataFrame({
    'A':['A3','A4','A5'],
    'B':['B3','B4','B5'],
    'C':['C3','C4','C5']
})
```

```python
# For Concatination we use function .concat()

# pd.concat(df1,df2) it will not work

pd.concat([df1,df2])
```

# Joining of 2 Dataframe

```python
df1 = pd.DataFrame({
    'name':['Alice','Bob','Charlie']    
},index=[1,2,3])
```

```python
df2 = pd.DataFrame({
    'score':[85,90,75]    
},index=[2,3,4])
```

```python
# For join 1 dataframe to other dataframe we use .join()

df1.join(df2)
```
