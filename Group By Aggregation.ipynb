```python
import numpy as np
import pandas as pd
```

```python
data = {
    'category':['A','B','A','A','B','A','B','B'],
    'stores':['S1','S1','S2','S2','S1','S2','S2','S1'],
    'sales':[100,200,150,250,120,180,200,300],
    'quantity':[10,15,12,18,8,20,15,25],
    'date':pd.date_range('2023-01-01',periods=8)
}
df = pd.DataFrame(data)
```

```python
df
```

# Group by 

use .groupby() when use it creates a object

Qualatative -> Categorical data (eg A,B or S1,S2)

Quantative -> (eg Stores , Category )

```python
# GroupBy Category & Calculate the sum of Sales

cat = df.groupby("category")["sales"].sum()
cat
```

```python
# GroupBy store & calculate the sum of sales

cat = df.groupby('stores')['sales'].sum()
cat
```

```python
# GroupBy Multiple Columns

cat = df.groupby(['category','sales'])['sales'].sum()
cat
```

# Aggregation

Multiple Elements combines together answer create

Mean , Std , Median , Sum , Variance , Min , Max , Count , (.agg)

```python
df['sales'].mean()
```

```python
df['sales'].mode()
```

```python
df['sales'].median()
```
